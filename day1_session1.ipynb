## Sentiment analysis using Transformer


```python
from transformers import pipeline

# Corrected the task name and input format
classifier = pipeline("sentiment-analysis")
results = classifier(["I have been waiting for this!", "I hate this so much!"])
print(results)
```

Text generation Model(like gemini)

```python
generator= pipeline("text-generation")
generator("once upon a time in nce robotics club...")
```

Language translator

```python
translator=pipeline("translation_en_to_de")
translator("I am a student")
```

qa pipeline

```python
from transformers import pipeline

qa = pipeline("question-answering")
qa(
    question="where am i?",
    context="once upon a time in nce robotics club...",
)
```

Ner Models

```python
ner= pipeline("ner",model="dbmdz/bert-large-cased-finetuned-conll03-english")
ner("I am Bishal From Lalitpur")
```
